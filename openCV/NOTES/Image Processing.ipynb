### flipping an image

```python
img = cv2.imread("my_car2.png")
horizontal = cv2.flip(img , 1)
vertical = cv2.flip(img , 0)
both = cv2.flip(img , -1)

plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(1,4,2)
plt.imshow(cv2.cvtColor(horizontal, cv2.COLOR_BGR2RGB))
plt.title("Horizontal")

plt.subplot(1,4,3)
plt.imshow(cv2.cvtColor(vertical, cv2.COLOR_BGR2RGB))
plt.title("Vertical")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(both, cv2.COLOR_BGR2RGB))
plt.title("Both")

plt.show()
```

#### cropping the images

```python
img.shape
```

```python
h , w = img.shape[:2]

# Define crop area safely inside image
row_start = int(h * 0.2)
row_end   = int(h * 0.6)
col_start = int(w * 0.2)
col_end   = int(w * 0.6)

cropped = img[row_start:row_end, col_start:col_end]

# Display
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title("Cropped")

plt.show()
```

### Transalation :

```python
img = cv2.imread('my_car.jpeg')

rows, cols = img.shape[:2]

M = np.float32([[1, 0, 150],
                [0, 1, -100]])
# 1st col - scaling , 2nd - slant  , 3rd for - shift 
translated = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.show()
```

### rotation of image

```python
center = (cols//2 , rows // 2)

M = cv2.getRotationMatrix2D(center , 30 , 1)

rotated = cv2.warpAffine(img , M , (cols,rows))
plt.imshow(cv2.cvtColor(rotated , cv2.COLOR_BGR2RGB))
plt.title("Rotated image")
plt.show()
```

### Drawing on image

     1. rectangle
     2. circle
     3. line
     4. text
    

```python
# Drawing Shapes on Images

import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('my_car.jpeg')

# Draw Rectangle
cv2.rectangle(
    img,
    (50, 50),      # top-left
    (300, 200),    # bottom-right
    (0, 255, 0),   # Green (BGR)
    3              # Thickness
)

# Draw Circle
cv2.circle(
    img,
    (500, 150),    # Center
    80,            # Radius
    (255, 0, 0),   # Blue (BGR)
    3              # Thickness
)

# Add Text
cv2.putText(
    img,
    "My Car",
    (300, 300),                    # Position
    cv2.FONT_HERSHEY_SIMPLEX,      # Font
    1,                             # Font Scale
    (0, 0, 255),                   # Red (BGR)
    2,                             # Thickness
    cv2.LINE_AA                    # Smooth Text
)

# Display Image
plt.figure(figsize=(10,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Rectangle + Circle + Text")
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

black = np.zeros((300,300), dtype=np.uint8)
dark = np.ones((300,300), dtype=np.uint8)
white = np.full((300,300), 255, dtype=np.uint8)
gray = np.full((300,300), 127, dtype=np.uint8)

plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.title("Gray")

plt.subplot(1,4,2)
plt.imshow(cv2.cvtColor(black, cv2.COLOR_BGR2RGB))
plt.title("Black")

plt.subplot(1,4,3)
plt.imshow(cv2.cvtColor(dark, cv2.COLOR_BGR2RGB))
plt.title("Dark-Almost Black")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(white, cv2.COLOR_BGR2RGB))
plt.title("White")

plt.show()
```

```python
# import cv2
import numpy as np

# ----------------------------
# Load Image
# ----------------------------
path = input("Enter image path: ")

img = cv2.imread(path)

if img is None:
    print("Image not found!")
    exit()

current = img.copy()


def show(image, title="Image"):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


while True:

    print("\n========== IMAGE EDITOR ==========")
    print("1. Display Image")
    print("2. Convert to Grayscale")
    print("3. Convert to HSV")
    print("4. Resize")
    print("5. Flip")
    print("6. Crop")
    print("7. Translate")
    print("8. Rotate")
    print("9. Draw Rectangle")
    print("10. Draw Circle")
    print("11. Add Text")
    print("12. Brightness")
    print("13. Contrast")
    print("14. Bitwise Circular Mask")
    print("15. Reset")
    print("16. Save")
    print("17. Exit")

    choice = int(input("Enter Choice : "))

    # -----------------------------------
    if choice == 1:

        show(current , "Image")

    # -----------------------------------
    elif choice == 2:

        gray = cv2.cvtColor(current, cv2.COLOR_BGR2GRAY)
        cv2.imshow("Gray", gray)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

    # -----------------------------------
    elif choice == 3:

        hsv = cv2.cvtColor(current, cv2.COLOR_BGR2HSV)
        cv2.imshow("HSV", hsv)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

    # -----------------------------------
    elif choice == 4:

        w = int(input("Width : "))
        h = int(input("Height: "))

        resized = cv2.resize(current, (w, h))

        show(resized , "Resized")

    # -----------------------------------
    elif choice == 5:

        print("0 = Vertical")
        print("1 = Horizontal")
        print("-1 = Both")

        flip = int(input())

        flipped = cv2.flip(current, flip)

        show(flipped , "image is flipped")

    # -----------------------------------
    elif choice == 6:

        x = int(input("Start X : "))
        y = int(input("Start Y : "))
        w = int(input("Width   : "))
        h = int(input("Height  : "))

        cropped = current[y:y+h, x:x+w]

        show(cropped , "cropped image")

    # -----------------------------------
    elif choice == 7:

        tx = int(input("Shift X : "))
        ty = int(input("Shift Y : "))

        M = np.float32([[1,0,tx],[0,1,ty]])

        translated = cv2.warpAffine(current, M,
                                 (current.shape[1],
                                  current.shape[0]))

        show(translated, "translated image")

    # -----------------------------------
    elif choice == 8:

        angle = float(input("Angle : "))

        h,w = current.shape[:2]

        center = (w//2,h//2)

        M = cv2.getRotationMatrix2D(center,
                                    angle,
                                    1)

        rotated = cv2.warpAffine(current,
                                 M,
                                 (w,h))

        show(rotated , "rotated image")

    # -----------------------------------
    elif choice == 9:

        x1 = int(input("x1 : "))
        y1 = int(input("y1 : "))
        x2 = int(input("x2 : "))
        y2 = int(input("y2 : "))
        rect = current.copy()

        cv2.rectangle(rect,
                      (x1,y1),
                      (x2,y2),
                      (0,255,0),
                      3)

        show(rect , "rectangle")

    # -----------------------------------
    elif choice == 10:
        circle = current.copy()
        x = int(input("Center X : "))
        y = int(input("Center Y : "))
        r = int(input("Radius : "))

        cv2.circle(circle,
                   (x,y),
                   r,
                   (255,0,0),
                   3)

        show(circle,"circle")

    # -----------------------------------
    elif choice == 11:
        text_img = current.copy()
        text = input("Text : ")

        x = int(input("X : "))
        y = int(input("Y : "))

        cv2.putText(text_img,
                    text,
                    (x,y),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    (0,0,255),
                    2)

        show(text_img,"text")

    # -----------------------------------
    elif choice == 12:

        beta = int(input("Brightness (-100 to 100): "))

        current5 = cv2.convertScaleAbs(current,
                                      alpha=1,
                                      beta=beta)

        show(current5,"brightness")

    # -----------------------------------
    elif choice == 13:

        alpha = float(input("Contrast (1-3): "))

        current6 = cv2.convertScaleAbs(current,
                                      alpha=alpha,
                                      beta=0)

        show(current6,"contrast")

    # -----------------------------------
    elif choice == 14:

        h,w = current.shape[:2]

        mask = np.zeros((h,w),
                        dtype=np.uint8)

        center = (w//2,h//2)

        radius = min(h,w)//3

        cv2.circle(mask,
                   center,
                   radius,
                   255,
                   -1)

        result = cv2.bitwise_and(current,
                                 current,
                                 mask=mask)

        cv2.imshow("Masked", result)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

    # -----------------------------------
    elif choice == 15:

        current7 = img.copy()

        print("Image Reset.")

    # -----------------------------------
    elif choice == 16:

        name = input("Output filename : ")

        cv2.imwrite(name,current)

        print("Image Saved.")

    # -----------------------------------
    elif choice == 17:

        print("Thank You!")
        break

    else:

        print("Invalid Choice")
```
